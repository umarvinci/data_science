# Popularity Based

It is a type of recommendation system which works on the principle of popularity and or anything which is in trend. These systems check about the product or movie which are in trend or are most popular among the users and directly recommend those.

For example, if a product is often purchased by most people then the system will get to know that that product is most popular so for every new user who just signed it, the system will recommend that product to that user also and chances becomes high that the new user will also purchase that. 

We will sort popularity of products based on whether they have been rated by the top users or not.

```python
df_user_rating.head()
```

```python
ratings_mean_count['rating_counts'] = pd.DataFrame(df_user_rating.groupby('productID')['ratings'].count())  
```

Since popularity based systems are not personalized, the system would recommend the same sort of products/movies which are solely based upon popularity to every other user. There is no personalisation based on user and every user would be recommended the same sort of products.

Now out of all products rated the most, we only recommend the ones which have been rated for more than 100 times so that there is no bias or a case of a product being popular because of one 5 star rating.

```python
df_subset = ratings_mean_count[ratings_mean_count['rating_counts']>100]
```

# Collaborative Filtering : SVD

To address some of the limitations of content-based filtering, __collaborative filtering__ uses similarities between users and items simultaneously to provide recommendations. This allows for serendipitous recommendations; that is, collaborative filtering models can recommend an item to user A based on the interests of a similar user B. Furthermore, the embeddings can be learned automatically, without relying on hand-engineering of features.



We make use of the Surprise library for Collaborative Filtering and the algorithms we will be using are SVD and KNN with Means.

```python
from surprise import SVD
from surprise import Dataset
from surprise.model_selection import cross_validate
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader
from collections import defaultdict
```

## Spliting data 70:30

```python
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df_user_rating, reader)
```

```python
trainset, testset = train_test_split(data, test_size=.30, random_state=42)
```

## fit algo on trainset (untuned)

```python
algo = SVD()
algo.fit(trainset)
```

## get predictions on testset

```python
predictions = algo.test(testset)
```

```python
predictions
```

Prediction of ratings for each product by each user. We use thiss value to calculate RMSE, our choice of metric for this project.

## Untuned SVD RMSE score

```python
print("Untuned SVD Model : Test Set")
accuracy.rmse(predictions, verbose=True)
```

## SVD Tuning : GridSearchCV

```python
from surprise.model_selection import GridSearchCV

param_grid = {'n_epochs': [5, 10], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)
gs.fit(data)
```

```python
print(f"Best score: {gs.best_score['rmse']}")
print(f"Best score: {gs.best_params['rmse']}")
```

## Tuned SVD Score

```python
tuned_algo = SVD(n_epochs= 10, lr_all= 0.005, reg_all= 0.4)
tuned_algo.fit(trainset)
```

```python
predictions = tuned_algo.test(testset)
print("Tuned SVD Model : Test Set")
accuracy.rmse(predictions, verbose=True)
```

```python
def get_top_n(predictions, n=5):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_n = get_top_n(predictions, n=5)
```

# Insights:

Both the algorithms, Single Value Decomposition and KNNwithMeans fared well for our recommendations.

There is no way of measuring the performance of popularity based recommendations so we have to take a guess at the threshold of minimum ratings and display the most rated products.

In collaborative filtering, we make use of SVD which is essentially a User-Item filter as well as User-user and item-item using KNN means

First we split the dataset into train and test set of 70:30 for both the algorithms. We then used the Surprise library for fitting the algorithms on train set and predicting on test set. Based on the predicted values, we evaluated our model based on RMSE values.

We make use of GridSearchCV to optimize the SVD algorithm. 

After having evaluated each model, we make use of another approach to show predictions. We split the whole(users with 50+ ratings only) into two parts,<br> train set: consisting of only ratings that has been provided by users, & <br> testset: which is the anti-train set consisting of all empty rating values.

We made use of this approach so that only products which have not been rated/purchased by users will be recommended to them based on their habits/habits of similar users. This ensured that no product which was rated high by the user shows up in the recommendations.

Lastly, we took two arbitraty users from the dataset and recommended the top 5 products personalised for them. This came out to be different for different users which makes this the best type of recommendation system in this project.

Since SVD had the lowest RMSE(0.99), we use that out of the two collaborative filtering models we had built to recommend products to our users.
